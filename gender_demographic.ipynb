```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'HUD_PIP_Data.csv'
df = pd.read_csv(file_path)
```

```python
# List of state codes you're interested in
state_codes = ['AZ', 'UT', 'CO', 'MA', 'LA', 'ND', 'MS', 'IL', 'NM', 'HI']

# Initialize empty lists to store total homeless counts for females and males
female_totals = []
male_totals = []

# Iterate over each state code
for state_code in state_codes:
    # Filter the DataFrame to get data only for the specified state and years from 2013 to 2023
    state_data = df[(df['State'] == state_code) & (df['Year'].between(2013, 2023))].copy()

    # Convert the "Overall Homeless - Female" and "Overall Homeless - Male" columns to numeric values, ignoring errors
    state_data['Overall Homeless - Female'] = pd.to_numeric(state_data['Overall Homeless - Female'], errors='coerce')
    state_data['Overall Homeless - Male'] = pd.to_numeric(state_data['Overall Homeless - Male'], errors='coerce')

    # Calculate total homeless counts for females and males for each state
    female_total = state_data['Overall Homeless - Female'].sum()
    male_total = state_data['Overall Homeless - Male'].sum()

    # Append total counts to the lists
    female_totals.append(female_total)
    male_totals.append(male_total)

# Plot bar graph to compare total homeless counts for females and males
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(state_codes))
plt.bar(index, female_totals, bar_width, label='Female')
plt.bar([i + bar_width for i in index], male_totals, bar_width, label='Male')

plt.xlabel('State')
plt.ylabel('Total Homeless Count')
plt.title('Comparison of Total Homeless Count by Gender')
plt.xticks([i + bar_width / 2 for i in index], state_codes)
plt.legend()
plt.tight_layout()
plt.show()
```
