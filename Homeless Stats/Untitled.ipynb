```python
import pandas as pd


# Read the CSV file into a DataFrame
df = pd.read_csv("/Users/syedshahid/Project 1/Project1_RealEstate/HUD_PIP_Data.csv")
```

```python
# Specify the columns of interest
columns_of_interest = [
    'Year',
    'State',
    'Overall Homeless - Under 18',
    'Overall Homeless - Age 18 to 24',
    'Overall Homeless - Over 24',
    'Overall Homeless - Female',
    'Overall Homeless - Male',
    'Overall Homeless - Transgender',
    'Overall Homeless - Gender that is not Singularly Female or Male',
    'Overall Homeless - Non-Hispanic/Non-Latin(o)(a)(x)',
    'Overall Homeless - Hispanic/Latin(o)(a)(x)',
    'Overall Homeless - White',
    'Overall Homeless - Black, African American, or African',
    'Overall Homeless - Asian or Asian American',
    'Overall Homeless - American Indian, Alaska Native, or Indigenous',
    'Overall Homeless - Native Hawaiian or Other Pacific Islander',
    'Overall Homeless - Multiple Races'
]
```

```python
# Specify the states of interest
states_of_interest = ['HI', 'UT', 'CO', 'MA', 'CA', 'LA', 'ND', 'MS', 'IL', 'NM', 'AZ']

# Filter the data to include only the specified columns, the years from 2013 to 2023, and the specified states
df_filtered = df[(df['Year'] >= 2013) & (df['Year'] <= 2023) & (df['State'].isin(states_of_interest))][columns_of_interest]

# Replace NaN values with 0
df_filtered.fillna(0, inplace=True)

# Remove the ".0" and round up the values
for col in df_filtered.columns[2:]:
    if pd.api.types.is_numeric_dtype(df_filtered[col]):
        df_filtered[col] = pd.to_numeric(df_filtered[col].astype(str).str.replace('.0', ''), errors='coerce').fillna(0).round().astype(int)

# Apply formatting
styled_df = df_filtered.style

# Set a caption
styled_df = styled_df.set_caption("Demographic Profiles for Selected States")

# Display the styled DataFrame
styled_df
```
