Ethnicity Homeless Demographic break down by state

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'HUD_PIP_Data.csv'
df = pd.read_csv(file_path)
```

```python
#All states sans California since they are just massive in comparison
# List of state codes you're interested in
state_codes = ['AZ', 'UT', 'CO', 'MA', 'LA', 'ND', 'MS', 'IL', 'NM', 'HI',]

# Iterate over each state code
for state_code in state_codes:
    # Filter the DataFrame to get data only for the specified state and years from 2013 to 2023
    state_data = df[(df['State'] == state_code) & (df['Year'].between(2013, 2023))].copy()

    # Convert the "Overall Homeless - Hispanic/Latin(o)(a)(x)" column to numeric values, ignoring errors
    state_data['Overall Homeless - Hispanic/Latin(o)(a)(x)'] = pd.to_numeric(state_data['Overall Homeless - Hispanic/Latin(o)(a)(x)'], errors='coerce')

    # Plot line chart for each state
    plt.plot(state_data['Year'], state_data['Overall Homeless - Hispanic/Latin(o)(a)(x)'], marker='o', linestyle='-', label=state_code)

plt.title('Overall Homeless - Hispanic/Latin(o)(a)(x) (2013-2023)')  # Title of the plot
plt.xlabel('Year')  # Label for the x-axis
plt.ylabel('Overall Homeless Hispanic/Latin(o)(a)(x)')  # Label for the y-axis
plt.grid(True)  # Show grid
plt.legend()  # Show legend with state codes
plt.tight_layout()  # Adjust layout
plt.show()  # Display the plot
```
